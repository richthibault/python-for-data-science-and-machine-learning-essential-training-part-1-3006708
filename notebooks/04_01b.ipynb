```python
import matplotlib as plt
import seaborn as sns

from pandas import DataFrame
```

```python
data = {'name':['John','Paul','George','Ringo'],
        'age':[24,24,22,26],
        'gender':['male','male','male','male'],
        'rank':[1,2,3,4]}
df = DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
plt.bar(df['name'], df['age'])
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Beatles by Age')
plt.show()
```

### Matplotlib's Bar Chart

```python
plot = sns.barplot(data=df, x='name', y='age')
plot.set_title('Beatle Ages')
plt.show()
```

### Line Plot Matplotlib

```python
plt.plot(df['name'], df['age'])
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Beatles by Age')
plt.show()
```

### Line Plot Seaborn

```python
plot = sns.lineplot(data=df, x='name', y='age')
plt.show()
```

### Pie Chart Matplotlib

```python
plt.pie(df['age'], labels=df['name'])
plt.title("Age Comparison")
plt.show()
```

### Pie Chart Seaborn

```python
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['age'], labels=df['name'], colors=colors)
plt.title("Age Comparison")
plt.show()
```
